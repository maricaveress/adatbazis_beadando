1. READ - temporary table létrehozása és adatokkal feltöltése

```python
import pyodbc

# Kapcsolat létrehozása az adatbázissal
conn = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=bit.uni-corvinus.hu;'
    'DATABASE=szallashely;'
    'UID=Hallgato;'
    'PWD=Password123;'
)

# Cursor objektum létrehozása SQL lekérdezésekhez
cursor = conn.cursor()

# READ - Temporary table létrehozása és adatokkal feltöltése
cursor.execute("SELECT * INTO #temp_szoba FROM szoba")

# Adatok kiírása
cursor.execute("SELECT * FROM #temp_szoba ")

# Output kiírása
print("Adatok:")
for row in cursor.fetchall():
    print(row)

# Kapcsolat bezárása
conn.close()
```

2. DELETE - A következőben törlöm az azt a sort az ideiglenes adatbázisból, aminek a szoba_id mezője 1 értéket vesz fel. 

```python
import pyodbc

# Kapcsolat létrehozása az adatbázissal
conn = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=bit.uni-corvinus.hu;'
    'DATABASE=szallashely;'
    'UID=Hallgato;'
    'PWD=Password123;'
)

# Cursor objektum létrehozása SQL lekérdezésekhez
cursor = conn.cursor()

# Temporary table létrehozása és adatokkal feltöltése
cursor.execute("SELECT * INTO #temp_szoba FROM szoba")

# Temporary table-ből egy sor törlése
cursor.execute("DELETE FROM #temp_szoba WHERE szoba_id = 1")


# Adatok kiírása
cursor.execute("SELECT * FROM #temp_szoba ")

# Output kiírása
print("Adatok:")
for row in cursor.fetchall():
    print(row)

# Kapcsolat bezárása
conn.close()
```

4. - UPDATE A művelet során módosítani fogom a 2-es szoba_id férőhely mezőjét.

```python
import pyodbc

# Kapcsolat létrehozása az adatbázissal
conn = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=bit.uni-corvinus.hu;'
    'DATABASE=szallashely;'
    'UID=Hallgato;'
    'PWD=Password123;'
)

# Cursor objektum létrehozása SQL lekérdezésekhez
cursor = conn.cursor()

# Temporary table létrehozása és adatokkal feltöltése
cursor.execute("SELECT * INTO #temp_szoba FROM szoba")

# Update - Egy sor frissítése a temporary table-ben.
cursor.execute("UPDATE #temp_szoba SET ferohely = 100 WHERE szoba_id = 2")

# Adatok kiírása
cursor.execute("SELECT * FROM #temp_szoba ")

# Output kiírása
print("Lekérdezés 1.")
for row in cursor.fetchall():
    print(row)

# Kapcsolat bezárása
conn.close()
```
