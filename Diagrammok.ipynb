```python
import pyodbc
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3

# Kapcsolat létrehozása az adatbázissal
conn = pyodbc.connect(
    'DRIVER={SQL Server};'
    'SERVER=bit.uni-corvinus.hu;'
    'DATABASE=szallashely;'
    'UID=Hallgato;'
    'PWD=Password123;'
)
# SQL lekérdezés definiálása
query = '''
        SELECT klimas, COUNT(*) AS count
        FROM szoba
        GROUP BY klimas
        '''

# Lekérdezés végrehajtása és eredmények lekérése
df = pd.read_sql(query, conn)

# Kapcsolat bezárása
conn.close()

# Adatok ábrázolása
plt.pie(df['count'], labels=df['klimas'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.title('Klímás szobák aránya')
plt.show()
```
